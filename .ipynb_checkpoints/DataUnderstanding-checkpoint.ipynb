```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("dataSet").getOrCreate()

file = "heart_2020_cleaned.csv"
df = spark.read.csv(file, header=True, inferSchema=True)

print("Shape:", (df.count(), len(df.columns)))

df.show()
```

```python
df.printSchema()
```

```python
numeric_features=['BMI', 'PhysicalHealth', 'MentalHealth', 'SleepTime']
categorical_features=['HeartDisease', 'Smoking', 'AlcoholDrinking', 'Stroke', 'DiffWalking', 'Sex', 'AgeCategory',
       'Race', 'Diabetic', 'PhysicalActivity', 'GenHealth','Asthma', 'KidneyDisease', 'SkinCancer']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

pandas_df = df.toPandas()

for feature in categorical_features:
    sns.set(style="whitegrid")
    g = sns.catplot(
        x=feature, hue="HeartDisease",
        data=pandas_df, kind="count", height=5, aspect=2
    )
    
    # Add numeric labels to each bar
    total_counts = pandas_df.groupby([feature])["HeartDisease"].count()
    for ax in g.axes.flat:
        for p in ax.patches:
            height = p.get_height()
            ax.annotate(
                f'{height}\n({height / total_counts.sum() * 100:.1f}%)', 
                (p.get_x() + p.get_width() / 2., height),
                ha='center', va='baseline', fontsize=12, color='black', xytext=(0, 5),
                textcoords='offset points'
            )
    
    # Setting up subgraph titles and labels
    g.set_axis_labels(feature, "Count")
    
    plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd

pandas_df = df.toPandas()


for feature in numeric_features:
    plt.figure(figsize=(8, 4))  
    sns.kdeplot(data=pandas_df, x=feature, hue='HeartDisease', fill=True, common_norm=False)

    plt.xlabel(feature)
    plt.ylabel("Densities")

    plt.grid(True)  

    plt.legend(title='HeartDisease', labels=['No', 'Yes']) 

    plt.show()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, sum



null_counts = df.select([sum(col(c).isNull().cast("int")).alias(c + "_null_count") for c in categorical_features+numeric_features])

null_counts.show()
```
